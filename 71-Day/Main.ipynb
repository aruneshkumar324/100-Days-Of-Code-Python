# Challenge

```python
# What college major has the highest mid-career salary?

index = new_df['Mid-Career Median Salary'].idxmax()
print(index)

college = new_df['Undergraduate Major'].loc[index]

print(college)

# How much do graduates with this major earn?

earn = new_df['Mid-Career Median Salary'].max()
print(earn)
```

```python
# Which college major has the lowest starting salary ?

index = new_df['Starting Median Salary'].idxmin()
print(index)

college = new_df['Undergraduate Major'].loc[index]
print(college)

earn = new_df['Starting Median Salary'].min()
print(earn)
```

```python
# Which college major has the lowest mid-career salary ?

index = new_df['Mid-Career Median Salary'].idxmax()
print(index)

college = new_df['Undergraduate Major'].loc[index]
print(college)

earn = new_df['Mid-Career Median Salary'].max()
print(earn)
```

# ADD NEW COLUMN

```python
# new_column = new_df['Mid-Career 90th Percentile Salary'].subtract(new_df['Mid-Career 10th Percentile Salary'])

new_column = new_df['Mid-Career 90th Percentile Salary'] - new_df['Mid-Career 10th Percentile Salary']

# print(new_column)

new_df.insert(1, 'Spread', new_column)

new_df.head()
```

# SORT VALUES

```python
sorted_df = new_df.sort_values('Spread')

sorted_df[['Undergraduate Major', 'Spread']]
```
